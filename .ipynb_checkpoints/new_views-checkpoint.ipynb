```python
import requests
import csv
import sys
```

```python
ORIGIN = "cs5700cdnorigin.ccs.neu.edu"

def get_content(port, path):
    try:
        url = "http://" + ORIGIN + ":" + str(port) + path
        req = requests.get(url).content
        return req
    except:
        return b''


result = dict()
```

```python
with open('pageviews.csv', newline='') as pages:
    lines = csv.reader(pages, delimiter=',')
    i = 0
    for line in lines:
        print(i)
        i += 1
        path = "/" + line[0]
        freq = int(line[1])
        c_size = sys.getsizeof(get_content(8080, path))
        result[path] = c_size * c_size * freq / (1*1024*1024 * 1024 * 1024)
```

```python
t = sorted(result.items(), key=lambda kv: (kv[1], kv[0]), reverse=True)
print(t)
```

```python
w_list = []
rank = len(t)

for item in t:
    temp = (item[0], rank)
    rank -= 1
    w_list.append(temp)

w_list
```

```python
total = 0
count = 0

for item in w_list:
    size = sys.getsizeof(get_content(8080, item[0]))
    total += size
    count += 1
    if total > 20*1024*1024:
        print(count)
        print(item[0])
        break
    
```
